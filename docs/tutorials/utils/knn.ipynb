# Building graphs

Using nearest neighbors algorithms on the [Iris dataset](https://en.wikipedia.org/wiki/Iris_flower_data_set).

```python
from IPython.display import SVG
```

```python
import pickle
```

```python
from sknetwork.embedding import GSVD
from sknetwork.utils import KNNDense, PNNDense
from sknetwork.visualization import svg_graph
```

## K-nearest neighbors

```python
iris = pickle.load(open('iris.p', 'rb'))
```

```python
data = iris['data']
```

```python
labels = iris['labels']
```

```python
knn = KNNDense(n_neighbors=3, undirected=True)
adjacency = knn.fit_transform(data)
```

```python
gsvd = GSVD(2, normalized=False)
position = gsvd.fit_transform(adjacency)
position = data[:,:2]
```

```python
image = svg_graph(adjacency, position, labels=labels, edge_weight=False)
```

```python
SVG(image)
```

## Parallel K-nearest neighbors

```python
pnn = PNNDense(n_neighbors=1, undirected=True)
adjacency = pnn.fit_transform(data)
```

```python
image = svg_graph(adjacency, position, labels=labels, edge_weight=False)
```

```python
SVG(image)
```
