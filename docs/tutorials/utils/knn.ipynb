# Building graphs

Using nearest neighbors algorithms on the [Iris dataset](https://en.wikipedia.org/wiki/Iris_flower_data_set).

```python
%matplotlib inline

import pickle
import networkx as nx
from sknetwork.utils import KNNDense, PNNDense
```

```python
COLORS = ['r', 'g', 'b']
```

```python
iris = pickle.load(open('iris.p', 'rb'))
```

### K-neighbors

```python
knn = KNNDense(n_neighbors=5, undirected=False)
adjacency = knn.fit_transform(iris['data'])
```

```python
nx.draw(nx.from_scipy_sparse_matrix(adjacency), node_color=[COLORS[y] for y in iris['labels']])
```

### Feature-wise K-neighbors

```python
fwknn = PNNDense(n_neighbors=2, undirected=True)
adjacency = fwknn.fit_transform(iris['data'])
```

```python
nx.draw(nx.from_scipy_sparse_matrix(adjacency), node_color=[COLORS[y] for y in iris['labels']])
```
