# Building graphs

Using nearest neighbors algorithms on vector data.

```python
%matplotlib inline

import pickle

import networkx as nx

from sknetwork.utils import KNeighborsTransformer, FWKNeighborsTransformer
```

```python
import pickle
```

```python
COLORS = ['r', 'g', 'b']
```

```python
iris = pickle.load(open('iris.p', 'rb'))
```

```python
iris
```

### K-neighbors

```python
knn = KNeighborsTransformer(n_neighbors=5, undirected=False)
adjacency = knn.fit_transform(iris['data'])
```

```python
nx.draw(nx.from_scipy_sparse_matrix(adjacency), node_color=[COLORS[y] for y in iris['target']])
```

### Feature-wise K-neighbors

```python
fwknn = FWKNeighborsTransformer(n_neighbors=2, undirected=True)
adjacency = fwknn.fit_transform(iris['data'])
```

```python
nx.draw(nx.from_scipy_sparse_matrix(adjacency), node_color=[COLORS[y] for y in iris['target']])
```
