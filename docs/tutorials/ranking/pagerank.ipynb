# PageRank

This notebook illustrates the ranking of the nodes of a graph by [PageRank](https://scikit-network.readthedocs.io/en/latest/reference/ranking.html#pagerank).

```python
from IPython.display import SVG
```

```python
import numpy as np
```

```python
from sknetwork.data import karate_club, painters, movie_actor
from sknetwork.ranking import PageRank, BiPageRank
from sknetwork.visualization import svg_graph, svg_digraph, svg_bigraph
```

## Graphs

**Loading**

```python
graph = karate_club(metadata=True)
adjacency = graph.adjacency
position = graph.position
```

**Ranking**

```python
pagerank = PageRank()
scores = pagerank.fit_transform(adjacency)
```

**Visualization**

```python
image = svg_graph(adjacency, position, scores=np.log(scores))
```

```python
SVG(image)
```

**Ranking with personalization**

```python
personalization = {1: 1, 10: 1}
```

```python
scores = pagerank.fit_transform(adjacency, personalization)
```

**Visualization**

```python
image = svg_graph(adjacency, position, scores=np.log(scores))
```

```python
SVG(image)
```

## Digraphs

**Loading**

```python
graph = painters(metadata=True)
adjacency = graph.adjacency
names = graph.names
position = graph.position
```

**Ranking**

```python
pagerank = PageRank()
scores = pagerank.fit_transform(adjacency)
```

**Visualization**

```python
image = svg_digraph(adjacency, position, scores=np.log(scores), names=names)
```

```python
SVG(image)
```

**Ranking with personalization**

```python
cezanne = 11
```

```python
scores = pagerank.fit_transform(adjacency, {cezanne:1})
```

**Visualization**

```python
image = svg_digraph(adjacency, position, scores=np.log(scores), names=names)
```

```python
SVG(image)
```

## Bigraphs

**Loading**

```python
graph = movie_actor(metadata=True)
biadjacency = graph.biadjacency
names_row = graph.names_row
names_col = graph.names_col
```

**Ranking**

```python
bipagerank = BiPageRank()
```

```python
drive = 3
aviator = 9
```

```python
bipagerank.fit(biadjacency, seeds={drive: 1, aviator: 1})
scores_row = bipagerank.scores_row_
scores_col = bipagerank.scores_col_
```

**Visualization**

```python
image = svg_bigraph(biadjacency, names_row, names_col, scores_row=np.log(scores_row), scores_col=np.log(scores_col))
```

```python
SVG(image)
```
